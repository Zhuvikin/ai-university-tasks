```python
import numpy as np

randomNumbers = np.random.randint(0, 101, size=100);

print(randomNumbers)
```

```python
import matplotlib.pyplot as plt

plt.plot(randomNumbers)
plt.show()
```

```python
# Python program for implementation of Bubble Sort
def bubbleSort(arr):
    n = len(arr)
 
    # Traverse through all array elements
    for i in range(n):
 
        # Last i elements are already in place
        for j in range(0, n-i-1):
 
            # traverse the array from 0 to n-i-1
            # Swap if the element found is greater
            # than the next element
            if arr[j] > arr[j+1] :
                arr[j], arr[j+1] = arr[j+1], arr[j]
 
# Driver code to test above 
bubbleSort(randomNumbers)

print(randomNumbers)

plt.plot(randomNumbers)
plt.show()
```

```python
x = np.arange(0, 100, 1)
y = np.random.randint(0, 101, 100)

fig, ax = plt.subplots()
ax.plot(x, y)

ax.set(xlabel='element num', ylabel='array values', title='Unsorted initial array')
ax.grid()
plt.show()
```

```python
ysorted = np.sort(y)
fig, ax = plt.subplots()
ax.plot(x, ysorted)

ax.set(xlabel='element num', ylabel='array values', title='Sorted array')
ax.grid()
plt.show()
```
